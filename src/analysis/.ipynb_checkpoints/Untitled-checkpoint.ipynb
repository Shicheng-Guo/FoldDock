```python
#Get separators and scores
X = newdf[['pLDDT',"numatoms","numres","pcd","mmd"]].values
y = newdf['DockQ'].values
    #Fit
kf = KFold(n_splits=5)

#5 fold CV
all_preds = []
all_true = []
for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    #reg = LinearRegression().fit(X_train, y_train)
    reg = RandomForestRegressor().fit(X_train, y_train)
    pred = reg.predict(X_test)
    pred[pred<0]=0
    pred[pred>1]=1
    #Save
    all_preds.extend([*pred])
    all_true.extend([*y_test])

#Array conversion
all_preds = np.array(all_preds)
all_true = np.array(all_true)
#Success overlap
frac_correct = np.intersect1d(np.argwhere(all_true>=0.23)[:,0],np.argwhere(all_preds>=0.23)[:,0]).shape[0]/np.argwhere(all_true>=0.23)[:,0].shape[0]
#Plot
R,p = pearsonr(all_preds,all_true)
err = np.average(np.absolute(all_preds-all_true))
fig,ax = plt.subplots(figsize=(12/2.54,12/2.54))
plt.scatter(all_preds, all_true,s=5,label='Pearson R='+str(np.round(R,2)),alpha=0.75)
plt.plot([0,1],[0,1],linestyle='--',color='gray')
plt.xlabel('Predicted DockQ')
plt.ylabel('True DockQ')
plt.title('5-fold CV results')
#Plot correct cutoff
plt.plot([0.23,0.23],[0,0.23],linewidth=1,linestyle='--',color='k')
plt.plot([0,0.23],[0.23,0.23],linewidth=1,linestyle='--',color='k',label='Success cutoff')
#Plot error regions
plt.plot([0.23,0.23],[0.23,1],linewidth=1,linestyle='--',color='r')
plt.plot([0.23,1],[0.23,0.23],linewidth=1,linestyle='--',color='r',label='error region')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.legend()
plt.tight_layout()
#plt.savefig(outdir+'rf-all.png',format='png',dpi=300)
#plt.close()
```
